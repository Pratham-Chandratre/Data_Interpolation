## Data Interpolation
Data interpolation is the process of estimating unknown values within a dataset based on the known values. In Python, there are various libraries available that can be used for data interpolation, such as NumPy, SciPy, and Pandas. Here is an example of how to perform data interpolation using the NumPy library:

1. Linear Interpolation
2. Cubic Interpolation With Scipy
3. Polynomial Interpolation

#  1.  Linear Interpolation

```python
# Works well with linear data
```

```python
import numpy as np

x = np.array([1,2,3,4,5])
y = np.array([2,4,6,8,10])
```

```python
import matplotlib.pyplot as plt

plt.scatter(x,y)
```

```python
# As we can see there are just 5 data points but if we want to add 10 more within this and following almost same relation we can use interpolation
```

```python
# As we can range of datapoints on x axis is from 1 to 5 so we need to add 10 datapoints between 1 to 5

X_new = np.linspace(1,5,10)  # It will create new x values

y_interpolation = np.interp(X_new,x,y)
```

```python
y_interpolation
```

```python
plt.scatter(y_interpolation,X_new)
```

# Cubic Interpolation With Scipy

```python
# Here our dataset will be in form of curves
```

```python
import numpy as np
x=np.array([1,2,3,4,5])
y=np.array([1,8,27,64,125]) # y is cube of x
```

```python
plt.scatter(x,y)
```

```python
from scipy.interpolate import interp1d  # 1d will be for 1 dimensional
```

```python
##create a cubic interpolation function

f = interp1d(x,y,kind='cubic')
```

```python
X_new = np.linspace(1,5,10)
y_interpolation = f(X_new)
```

```python
y_interpolation
```

```python
plt.scatter(X_new, y_interpolation)
```

# 3. Polynomial Interpolation

```python
import numpy as np

# create some sample data
x = np.array([1, 2, 3, 4, 5])
y = np.array([1, 4, 9, 16, 25])  # y is square of x
```

```python
plt.scatter(x,y)
```

```python
# interpolate the data using polynomial interpolation
p=np.polyfit(x,y,2)  # If we give degree 1 it will be linear, if 2 some exponential if 3 more exponential and so on.
```

```python
X_new = np.linspace(1,5,10)
y_interpolate = np.polyval(p, X_new)
```

```python
plt.scatter(X_new, y_interpolate)
```
